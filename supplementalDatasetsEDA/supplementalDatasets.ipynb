## Allergens Basic EDA

```python
# Conducting basic EDA for allergies_10k.csv in the same folder
allergies_df = pd.read_csv('./allergies_10k.csv')
print(allergies_df.info())
print(allergies_df.describe())
```

```python
allergies_df.head()
```

```python
allergies_df['main_ingredient'] = allergies_df['ingredient'].str.split('(').str[0]
allergies_df.head()
```

```python
allergens = [allergen for allergen in allergies_df['allergens'] if allergen != '[]']

# the string for each item in allergens is a list of allergens, so we need to split the string and get the unique allergens
allergen_list = []
for allergen in allergens:
    allergen_list.extend(allergen.strip('[]').split(', '))
unique_allergens = list(set([allergen.strip("'").strip("**") for allergen in set(allergen_list)]))
unique_allergens
```

```python
# creating a df with the allergies_df['main_ingredient'] as the indices and the unique allergens as columns, and the values are 1 if the main ingredient is in the allergen list for that row, and 0 otherwise
allergen_df = pd.DataFrame(columns=unique_allergens, index=allergies_df['main_ingredient'])
for index, row in allergies_df.iterrows():
    main_ingredient = row['main_ingredient']
    allergens = row['allergens']
    if allergens != '[]':
        allergen_list = allergens.strip('[]').split(', ')
        for allergen in allergen_list:
            allergen_df.loc[main_ingredient, allergen.strip("'").strip("**")] = 1
allergen_df = allergen_df.fillna(0)
allergen_df.head()
```

```python
allergen_counts = allergen_df.sum()
allergen_counts
```

```python
# graph the number of main ingredients that have each allergen
allergen_counts.plot(kind='bar', figsize=(12, 6), title='Number of Main Ingredients with Each Allergen', xlabel='Allergen', ylabel='Count', rot=45)
```
